```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

seed = 2
np.random.seed(seed)
```

```python
from sklearn.preprocessing import LabelEncoder
# Loading the telescope data
# We skip the first row, as it consists of labels
# We specify that the data is seperated by commas(",") 
# We specify our datatype as a string
dataset = np.loadtxt("gamma-telescope-data.csv", skiprows=1, delimiter=",", dtype=str)
# We split the telescope data into input and output variables
# Our input will be everything between the 0th column & the 10th column
X = dataset[:,0:10]
# Converted the np array that carried the datatype string to a np array that consists of float values.
# To aid in neural net calculations ahead.
X = X.astype(np.float)
# Our output will be everything after the 10th column.
Y = dataset[:,10]

# Converting the output values that are g's and h's to integers.
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y=encoder.transform(Y)

# print "Input",X
# print len(X)
print "Ouptut",encoded_Y
```

```python
# Creating Layers one at a time because networks in Keras are defined as a sequence of layers.
# The first thing to get right is the number of inputs. in our case 10.

# Next, in order to determine how many layers are needed and the types of layers
# , is a matter of experimentation(trial and error.)

# For this problem, I will try to use a fully connected network with three layers.
# Fully connected layers are defined using the "Dense" class.
model = Sequential()
model.add(Dense(12, input_dim=10, init='uniform', activation='relu'))
model.add(Dense(10, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
model.fit(X,encoded_Y, nb_epoch=150, batch_size=10)
```
